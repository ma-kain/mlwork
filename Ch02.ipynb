```r
#generate a vector of random normal variables
#mean = 0 and stdev = 1
x = rnorm(n=10)
x
y = x + rnorm(n=10, mean=0, sd=0.1)
#Find correlation between variables
cor(x,y)
```

<b>Contour plot

```r
x = 1:50
y = seq(from=1,to=100,by=2)
#create product of vectors and their result
z = outer(x,y,function(x,y)sqrt(x^2+y^2))
contour(x,y,z)
contour(x,y,z, nlevels=25,add=T)
```

```r
#Create Color-coded plot, known as heatmap
image(x,y,z)
```

```r
#Produce 3-d plot
persp(x,y,z)
```

```r
#theta and phi control the viewing angle
par(mfrow=c(2,2)) 
persp(x,y,z, theta=0, phi=0)
#rotate 30deg left
persp(x,y,z, theta=30, phi=0)
#rotate 30deg right
persp(x,y,z, theta=-30, phi=0)
```

```r
par(mfrow=c(2,2)) 
persp(x,y,z, theta=0, phi=0)
#rotate 30deg towards front
persp(x,y,z, theta=0, phi=30)
#rotate 30deg towards back
persp(x,y,z, theta=0, phi=-30)
```

<b>Loading Data

```r
#Importing data set into R
D = read.table("Auto.data")
dim(D)
```

```r
#Import CSV data
#use first line as header
#consider ? mark as NA means missing element
D = read.csv("Auto.csv", header = T, na.strings='?')
dim(D)
```

```r
#remove rows with NA elements
D = na.omit(D)
dim(D)
```

```r
#return column names of data set
names(D)
```

<b>Additional Graphical and Numerical Summaries

```r
#plot scatter plot between two variables of data set
plot(D$cylinders,D$mpg)

#variables are not accessible without D$ prefix
#to make data set variables with by name use attach
#attach(D) #run just once
print(mpg[1:5])
```

```r
#cylinders is numeric vector, quantitative variable
summary(cylinders)

#convert to qualitative varibale
cyl = as.factor(cylinders)
print(summary(cyl))
```

<b>Box plot

```r
#Box plot automatically produced for qualitative variable in x-axis
#enable variable width to represent population in the category
#color added
plot(cyl,mpg, varwidth=T, col="red")
```
