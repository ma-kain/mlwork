```r
set.seed(1)
```

<b>(a) Using the rnorm() function, create a vector, x, containing 100 observations drawn from a N (0, 1) distribution. This represents a feature, X.

```r
x=rnorm(100, mean = 0, sd = 1)
```

<b>(b) Using the rnorm() function, create a vector, eps, containing 100 observations drawn from a N(0,0.25) distribution i.e. a normal distribution with mean zero and variance 0.25.

```r
eps=rnorm(100, mean = 0, sd = 0.25)
```

<b>(c) Using x and eps, generate a vector y according to the model Y =−1+0.5X+ε. (3.39)
What is the length of the vector y? What are the values of β0 and β1 in this linear model?</b><br>
> <font color="green">
    Comment: β0 = -1 and β1 = 0.5
    </font>

```r
y= -1 + 0.5 * x + eps
```

<b>(d) Create a scatterplot displaying the relationship between x and y. Comment on what you observe.</b><br>
> <font color="green">
    Comment: Data point density is higher towards the x-mean and grandually decreases away from the x-mean.
    </font>

```r
plot(x,y)
```

<b>(e) Fit a least squares linear model to predict y using x. Comment on the model obtained. How do βˆ0 and βˆ1 compare to β0 and β1?</b><br>
> <font color="green">
    Comment: Estimated coefficients are quite matching with true ones. βˆ0 = -1.01 and βˆ1 = 0.49 
    </font>

```r
lmodel=lm(y~x)
summary(lmodel)
```

<b>(f) Display the least squares line on the scatterplot obtained in (d). Draw the population regression line on the plot, in a different color. Use the legend() command to create an appropriate legend.

```r
plot(x,y)
abline(lmodel, col="red", pch=17, lwd=3)
abline(-1, 0.5, col="green", pch=19, lwd=2)
legend("topleft", legend=c("model","true"), col=c("red","green"), pch = c(19))
```

<b>(g) Now fit a polynomial regression model that predicts y using x and x2. Is there evidence that the quadratic term improves the model fit? Explain your answer.</b><br>
> <font color="green">
    Comment: Error and R-squared has improved slightly.
    </font>

```r
lm.poly = lm(y~x+I(x^2))
summary(lm.poly)
```

<b>(h) Repeat (a)–(f) after modifying the data generation process in such a way that there is less noise in the data. The model (3.39) should remain the same. You can do this by decreasing the variance of the normal distribution used to generate the error term ε in (b). Describe your results.</b><br>
> <font color="green">
    Comment: Seen improvement in Standard error and R-squared value after reducing noise in eps.
    </font>

```r
x=rnorm(100, mean = 0, sd = 1)
eps=rnorm(100, mean = 0, sd = 0.1)
y= -1 + 0.5 * x + eps
lmodel=lm(y~x)
summary(lmodel)
plot(x,y)
abline(lmodel, col="red", pch=17, lwd=3)
abline(-1, 0.5, col="green", pch=19, lwd=2)
legend("topleft", legend=c("model","true"), col=c("red","green"), pch = c(19))
```
